#### Looking for Categorical Data

```python
categorical_columns = []
for column in df.columns :
  if column not in df._get_numeric_data().columns :
    categorical_columns.append(column)
categorical_columns
```

## Converting Categorical Values into Numerical

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])
```

## Generating Corelation Matrix

```python
corelation_matrix = df.corr()
sns.heatmap(corelation_matrix, cmap=plt.cm.Reds, annot=True)
plt.show()
```

## Splitting Data into Features and Target

```python
x = df.iloc[: , : -1]
y = df.iloc[: , -1]
```

```python
x
```

```python
y
```

## Split Data into Training and Testing

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## Training Model

```python
scale = preprocessing.StandardScaler()
```

```python
X_trainscaled = scale.fit_transform(x_train)
clf = MLPClassifier(activation='relu', solver = 'adam', early_stopping = True, hidden_layer_sizes=(128,64,32), verbose = 1, validation_fraction=0.2, max_iter = 100)
clf.fit(X_trainscaled, y_train)
```

## Testing the Model

```python
X_testscaled = scale.transform(x_test)
y_pred = clf.predict(X_testscaled)
test_accuracy = metrics.accuracy_score(y_test, y_pred)
print("Testing Accuracy:", test_accuracy)
```
